# Loading data

```python
import pandas as pd

# Raw URLs for CSV files
url_inventory = 'https://raw.githubusercontent.com/b-ukusi/datasets/main/Resources%20Inventory%20Cost%20Sheet.csv'
url_cyst = 'https://raw.githubusercontent.com/b-ukusi/datasets/main/Ovarian%20Cyst%20Track%20Data.csv'
df_inventory = pd.read_csv(url_inventory)
df_cyst      = pd.read_csv(url_cyst)

# ðŸ‘‡ Here: drop columns that are entirely empty (no data in any row)
df_inventory = df_inventory.loc[:, df_inventory.notna().any(axis=0)]
df_cyst      = df_cyst.loc[:,      df_cyst.notna().any(axis=0)]

# Configure full display
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)


# Show data
print("Inventory data:")
display(df_inventory.head(100))

print("\nCyst data:")
display(df_cyst.head(100))
```
